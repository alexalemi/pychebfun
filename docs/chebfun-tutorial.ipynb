Chebfun tutorial================Welcome to chebfun.  It is meant to be a python module that mimics the matlab project: [Chebfun](http://www2.maths.ox.ac.uk/chebfun/)The aim is to create an environment for easy "analytical" exploration:  Chebfun is to analytics what floating point is to arithmetic

Basic usage-----------The library works well as a "calculator" for an ipython notebook session, so you can import *

```python
#from pychebfun import *   #if installed#for now lets run run it locally%run ../pychebfun.py
```

Where you import star, you'll get $x$, which is a simple chebfun defined on the interval (-1,1), as well as some overloaded math operators, so you can just go about your day for the most partIf you ask for a function of a chebfun, you'll get a new chebfun

```python
cos(x)
```

You can then create more or less arbitrary functions of this chebfun, and you will get new chebfuns

```python
f = 1./(1+x**2)*cos(5*x)f
```

The chebfun is a Chebyshev polynomial approximation of the function you've asked to compute.  Asking for its length tells you how many points are currently being used

```python
len(f)
```

You can see the coefficients themselves as well

```python
print f.coef
```

These internally, the algorithm uses higher and higher order interpolations, until the coefficients become smallYou're chebfun also behaves like a numpy ufunc, giving you near machine precision values at any point

```python
func = lambda x: 1./(1+x**2)*sin(5*x)f = Chebfun(func)pts = rand(10)real = func(pts)print f(pts)print realprint f(pts)-real
```

You can see a plot of the error on a regular grid with .coefplot()

```python
f.errplot()
```

The errors should *hopefully* always be near machine precision on the interval

You can also take derivatives and integrals of your function with ease

```python
display(f.deriv())display(f.integ())
```

You can differentiate to any order

```python
f.deriv(5)
```

And Integrate to any order, and can set the lower bound

```python
f.integ(5,lbnd=-1)
```

You also have access to the real roots of your function, at .introots()

```python
roots = f.introots()print "roots @ ", rootsdisplay(f)inflections = f.deriv(2).introots()plot(roots,zeros_like(roots),"rx")
```

As well as the range, computed from the extremums from the derivative of the chebfun

```python
display(f)print f.range
```

You can also mix and match different Chebfuns

```python
g = cos(5*x)h = 1./(1+x**2)display(g)display(h)display(g/h)
```

Defining Chebfuns------------------You can define Chebfuns in other ways, either with a python callable, or a string, evaluated in the numpy namespace

```python
f = Chebfun(lambda x: exp(-x**2))g = Chebfun("exp(-x**2)")display(f)display(g)
```

The second argument to Chebfun gives a domain the function is defined on

```python
f = Chebfun("sin(x)",(0,3*pi))display(f)
```

```python
f = Chebfun("sin(x)+sin(x**2)",(0,10))f
```

A good way to work is to start off defining $x$ on your range and going from there, as composing chebfuns keeps the rangeA convenience method `xdom` exists to facilitate the process

```python
xp = xdom((0,10))display(sin(xp))display(cos(xp))
```
